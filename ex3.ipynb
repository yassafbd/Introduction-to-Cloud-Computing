```python
from firebase import firebase
FBconn = firebase.FirebaseApplication('FIREBASE_URL',None)

def add_single_word(word):
    word = word.lower().strip()
    current_count = FBconn.get('/words', word)
    if current_count is None:
        new_count = 1
    else:
        new_count = current_count + 1
    FBconn.put('/words', word, new_count)

def add_text_analysis(text):
    clean_text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    words = clean_text.lower().split()
    for word in words:
        add_single_word(word)

def update_word_count(word, count):
    FBconn.put('/words', word, count)

def delete_word(word):
    FBconn.delete('/words', word)

def view_all_words():
   result = FBconn.get('/words', None)
   if result:
       output = [f"{word}: {count}" for word, count in result.items()]
       print(", ".join(output))

```

```python
while True:
  print("\nCommon Words Counter Menu:")
  print("1. Add single word")
  print("2. Add text for analysis")
  print("3. Update word count")
  print("4. Delete word")
  print("5. View all words")
  print("6. Exit")

  choice = input("\nSelect an option (1-6): ")

  if choice == '1':
        w = input("Enter word: ")
        add_single_word(w)
  elif choice == '2':
        t = input("Enter text: ")
        add_text_analysis(t)
  elif choice == '3':
        w = input("Enter word to update: ")
        c = int(input("Enter new count: "))
        update_word_count(w, c)
  elif choice == '4':
        w = input("Enter word to delete: ")
        delete_word(w)
  elif choice == '5':
        view_all_words()
  elif choice == '6':
        print("Goodbye!")
        break
  else:
        print("Invalid choice, try again.")
```

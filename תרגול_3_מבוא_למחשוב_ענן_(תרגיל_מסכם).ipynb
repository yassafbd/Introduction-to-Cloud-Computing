```python
def add_single_word():
    # 1. קליטת המילה מהמשתמש וניקוי רווחים
    word = input("Enter a word: ").lower().strip()

    if not word:
        print("Invalid input.")
        return

    # 2. ניסיון לקרוא את הערך הנוכחי של המילה מתוך התיקייה 'words'
    # אנחנו משתמשים במילה עצמה כמפתח (Key) בתוך התיקייה
    current_count = FBconn.get('/words/', word)

    if current_count is None:
        # אם המילה לא קיימת, נגדיר מונה חדש כ-1
        new_count = 1
    else:
        # אם המילה קיימת, נוסיף 1 למונה הקיים
        new_count = current_count + 1

    # 3. שמירת הערך המעודכן ב-Firebase (שימוש ב-put כדי לעדכן לפי מפתח ספציפי)
    FBconn.put('/words/', word, new_count)

    print(f"The word '{word}' was updated. New count: {new_count}")
```

```python
def add_text_analysis():
    # 1. קליטת טקסט מהמשתמש
    text = input("Enter text for analysis: ").lower()

    # 2. פירוק הטקסט לרשימת מילים (באמצעות split)
    words_list = text.split()

    if not words_list:
        print("No words found in text.")
        return

    print(f"Analyzing {len(words_list)} words...")

    # 3. מעבר על כל מילה ברשימה
    for word in words_list:
        # ניקוי סימני פיסוק בסיסיים אם יש (אופציונלי)
        word = word.strip('.,!?;:"()')

        if word:
            # בדיקה מול Firebase (בדיוק כמו בפונקציה הקודמת)
            current_count = FBconn.get('/words/', word)

            if current_count is None:
                new_count = 1
            else:
                new_count = current_count + 1

            # עדכון המונה ב-Firebase
            FBconn.put('/words/', word, new_count)

    print("Analysis complete. Database updated.")
```

```python
def update_word_count():
    # 1. קבלת המילה והמונה החדש מהמשתמש
    word = input("Enter the word you want to update: ").lower().strip()

    # בדיקה אם המילה בכלל קיימת בבסיס הנתונים (אופציונלי אך מומלץ)
    current_val = FBconn.get('/words/', word)

    if current_val is None:
        print(f"The word '{word}' does not exist in the database.")
        # אפשר לשאול את המשתמש אם הוא רוצה ליצור אותה בכל זאת, או פשוט לצאת
        return

    try:
        new_count = int(input(f"Current count for '{word}' is {current_val}. Enter new count: "))

        # 2. עדכון הערך ב-Firebase באמצעות put
        # פקודה זו ניגשת לנתיב /words/word ומחליפה את הערך שם ב-new_count
        FBconn.put('/words/', word, new_count)

        print(f"Successfully updated '{word}' to {new_count}.")
    except ValueError:
        print("Invalid input. Please enter a number for the count.")
```

```python
def delete_word():
    # 1. קבלת המילה למחיקה מהמשתמש
    word = input("Enter the word you want to delete: ").lower().strip()

    # בדיקה אם המילה קיימת לפני הניסיון למחוק (אופציונלי)
    current_val = FBconn.get('/words/', word)

    if current_val is None:
        print(f"The word '{word}' was not found in the database.")
        return

    # 2. ביצוע המחיקה מתוך התיקייה 'words'
    # פקודת delete מסירה את המפתח ואת הערך שלו לצמיתות [cite: 323]
    FBconn.delete('/words/', word)

    print(f"The word '{word}' has been deleted from the database.")
```

```python
def view_all_words():
    # 1. משיכת כל הנתונים מתיקיית ה-'words'
    # פקודת get עם None מחזירה את כל התוכן של הנתיב כמילון (Dictionary)
    all_words = FBconn.get('/words/', None)

    # 2. בדיקה האם יש נתונים במסד הנתונים
    if all_words:
        print("\n--- Current Word Counts ---")

        # 3. מעבר על המילון והדפסת כל מילה (key) והמונה שלה (value)
        # אנחנו משתמשים ב-items() כדי לקבל גם את המפתח וגם את הערך בכל סיבוב
        for word, count in all_words.items():
            print(f"{word} : {count}")
    else:
        # טיפול במקרה שבו מסד הנתונים ריק
        print("\nNo words found in the database.") [cite: 334-335]
```

```python
import pandas as pd
import altair as alt

def show_word_chart():
    all_words = FBconn.get('/words/', None)
    if not all_words:
        print("No data available.")
        return

    # הפיכה לטבלה
    data = [{'Word': w, 'Count': c} for w, c in all_words.items()]
    df = pd.DataFrame(data)

    # יצירת הגרף
    chart = alt.Chart(df).mark_arc().encode(
        theta=alt.Theta(field="Count", type="quantitative"),
        color=alt.Color(field="Word", type="nominal"),
        tooltip=['Word', 'Count']
    ).properties(title='Word Frequency')

    display(chart)
```
